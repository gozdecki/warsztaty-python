```python
for i in range(1,11):
    for j in range(1,11):
        if i*j < 10:
            print(str(j*i) + '  ', end=' ')
        if (i*j < 100) and (i*j > 9):
            print(str(j*i) + ' ', end=' ')
        if (i*j < 1000) and (i*j > 99):
            print(str(j*i) + '', end=' ')
    print('')    
        
```
